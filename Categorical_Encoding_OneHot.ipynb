FROM SKLEARN

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
```

```python
usecols = ["Pclass", "SibSp", "Parch", "Sex", "Embarked", "Cabin", "Survived"]
data = pd.read_csv("Titanic-Dataset.csv",usecols=usecols)
```

```python
data["Cabin"]=data["Cabin"].str[0]
```

```python
data.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    data.drop("Survived",axis=1),
    data["Survived"],
    test_size=0.3,
    random_state=0
)
```

```python
x_train.shape,x_test.shape
```

**Limitations**

1)need to set specific output to return pandas

2)need additional transformer to encode variable subset

3)changes variable names after transformation

```python
x_train.fillna("missing",inplace=True)
x_test.fillna("missing",inplace=True)
```

```python
encoder= OneHotEncoder(
    categories="auto",
    drop="first",
    sparse_output=False,
    handle_unknown="error"
)
encoder.set_output(transform="pandas")
```

```python
encoder.fit(x_train)
```

```python
encoder.categories_
```

```python
x_train_enc = encoder.transform(x_train)
x_test_enc = encoder.transform(x_test)
```

```python
x_train_enc.head()
```

```python
x_test_enc.head()
```

```python
encoder.get_feature_names_out()
```

```python
ct=ColumnTransformer([
    ("encoder",encoder,["Sex", "Embarked", "Cabin"])],remainder="passthrough")
ct.set_output(transform="pandas")
```

```python
ct.fit(x_train)
```

```python
x_train_enc= ct.transform(x_train)
x_test_enc=ct.transform(x_test)
```

```python
x_train_enc.head()
```

```python
# USE PIPELINE
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    data.drop("Survived",axis=1),
    data["Survived"],
    test_size=0.3,
    random_state=0
)
```

```python
x_train.shape,x_test.shape
```

```python
encoder = OneHotEncoder(
    categories="auto",
    drop="first",
    sparse_output=False,
    handle_unknown="error"
)
```

```python
pipe = Pipeline([
    ("imputer",SimpleImputer(strategy="constant", fill_value="missing")),
    ("ohe",encoder)
])
```

```python
ct = ColumnTransformer([
    ("encoder", pipe, ["Sex", "Embarked", "Cabin"])], remainder="passthrough")
```

```python
ct.set_output(transform="pandas")
```

```python
ct.fit(x_train)
```

```python
x_train_enc=ct.transform(x_train)
x_test_enc=ct.transform(x_test)
```

```python
x_train_enc.shape,x_test_enc.shape
```

**USING FEATURE ENGINE**

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline

# Feature-engine
from feature_engine.encoding import OneHotEncoder
from feature_engine.imputation import CategoricalImputer
```

```python
usecols = ["Pclass", "SibSp", "Parch", "Sex", "Embarked", "Cabin", "Survived"]

data = pd.read_csv("Titanic-Dataset.csv", usecols=usecols)
data["Cabin"] = data["Cabin"].str[0]

data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data.drop("Survived", axis=1),
    data["Survived"],
    test_size=0.3,
    random_state=0,
)
X_train.shape, X_test.shape
```

```python
X_train.fillna("Missing", inplace=True)
X_test.fillna("Missing", inplace=True)
```

```python
encoder = OneHotEncoder(
    variables=None,
    drop_last=True,
)
```

```python
encoder.fit(X_train)
```

```python
encoder.variables_
```

```python
X_train_enc = encoder.transform(X_train)
X_test_enc = encoder.transform(X_test)

X_train_enc.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data.drop("Survived", axis=1),  # predictors
    data["Survived"],
    test_size=0.3,
    random_state=0,
)
X_train.shape, X_test.shape
```

```python
X_train.isnull().sum()
```

```python
pipe = Pipeline(
    [
        ("imputer", CategoricalImputer()),
        ("ohe", OneHotEncoder(drop_last=True)),
    ]
)
```

```python
pipe.fit(X_train)
```

```python
X_train_enc = pipe.transform(X_train)
X_test_enc = pipe.transform(X_test)
X_train_enc.head()
```

```python
X_train_enc.isnull().sum()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from category_encoders.one_hot import OneHotEncoder
```

**One hot encoding with Category encoders**

**Advantages:**

quick

returns dataframe

returns feature names

allows to select features to encode

appends dummies to original dataset

**Limitations**:

No option for k-1 dummies

```python
X_train, X_test, y_train, y_test = train_test_split(
    data.drop("Survived", axis=1),  # predictors
    data["Survived"],  # target
    test_size=0.3,  # percentage of obs in test set
    random_state=0, # seed to ensure reproducibility
)

X_train.shape, X_test.shape
```

```python
X_train.fillna("Missing", inplace=True)
X_test.fillna("Missing", inplace=True)
```

```python
encoder = OneHotEncoder(
    cols=None,  # alternatively pass a list of variables
    use_cat_names=True,
)
```

```python
# fit the encoder (finds categories)

encoder.fit(X_train)
```

```python
encoder.cols
```

```python
encoder.mapping
```

```python
X_train_enc = encoder.transform(X_train)
X_test_enc = encoder.transform(X_test)

X_train_enc.head()
```
