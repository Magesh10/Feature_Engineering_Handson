USING PANDAS(Manualy create)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("House_price.csv",usecols=["Neighborhood", "Exterior1st", "Exterior2nd", "SalePrice"]
)
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data.drop("SalePrice", axis=1),
    data["SalePrice"],
    test_size=0.3,
    random_state=0,
)
X_train.shape, X_test.shape
```

```python
#first let's create a dictionary with the mappings of categories to numbers
ordinal_maping = {k:i for i,k in enumerate(X_train["Neighborhood"].unique(),0)}
```

```python
ordinal_maping
```

```python
# replace the labels with the integers

X_train["Neighborhood"] = X_train["Neighborhood"].map(ordinal_maping)
X_test["Neighborhood"] = X_test["Neighborhood"].map(ordinal_maping)
```

```python
X_train["Neighborhood"].head(10)
```

```python
def find_category_mappings(df, variable):
    return {k: i for i, k in enumerate(df[variable].unique(), 0)}

def integer_encode(train, test, variable, ordinal_mapping):

    X_train[variable] = X_train[variable].map(ordinal_mapping)
    X_test[variable] = X_test[variable].map(ordinal_mapping)
```

```python
for variable in ["Exterior1st", "Exterior2nd"]:
    mappings = find_category_mappings(X_train, variable)
    integer_encode(X_train, X_test, variable, mappings)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = pd.read_csv("House_price.csv",usecols=["Neighborhood", "Exterior1st", "Exterior2nd", "SalePrice"]
)
data.head()
```

```python
for cols in data.columns:
  print(cols,": ",len(data[cols].unique())," lables")
```

```python
data["Neighborhood"].unique()
```

```python
data["Exterior1st"].unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data.drop("SalePrice", axis=1),
    data["SalePrice"],
    test_size=0.3,
    random_state=0,
)

X_train.shape, X_test.shape
```

```python
X_train.isnull().sum()
```

Ordinal Encoding with Scikit-learn

```python
cat_var = list(X_train.select_dtypes(include="O").columns)
```

```python
cat_var
```

```python
encoder = OrdinalEncoder()
```

```python
ct = ColumnTransformer(
    [("ordinalencoder", encoder, cat_var)],
    remainder="passthrough",
)
ct.set_output(transform="pandas")
```

```python
ct.fit(X_train)
```

```python
X_train_enc = ct.transform(X_train)
X_test_enc = ct.transform(X_test)

X_train_enc.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

from feature_engine.encoding import OrdinalEncoder
```

```python
data = pd.read_csv("House_price.csv",usecols=["Neighborhood", "Exterior1st", "Exterior2nd", "SalePrice"]
)
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data[["Neighborhood", "Exterior1st", "Exterior2nd"]],
    data["SalePrice"],
    test_size=0.3,
    random_state=0,
)

X_train.shape, X_test.shape
```

```python
ordinal_enc = OrdinalEncoder(
    encoding_method="arbitrary",
    variables=["Neighborhood", "Exterior1st", "Exterior2nd"])

ordinal_enc.fit(X_train)
```

```python
ordinal_enc.encoder_dict_
```

```python
X_train = ordinal_enc.transform(X_train)
X_test = ordinal_enc.transform(X_test)

X_train.head()
```

Count or frequency encoding - Feature-engine

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# to encode with feature-engine
from feature_engine.encoding import CountFrequencyEncoder
```

```python
data = pd.read_csv("House_price.csv",usecols=["Neighborhood", "Exterior1st", "Exterior2nd", "SalePrice"]
)
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data[["Neighborhood", "Exterior1st", "Exterior2nd"]],
    data["SalePrice"],
    test_size=0.3,
    random_state=0,
)

X_train.shape, X_test.shape
```

```python
count_enc = CountFrequencyEncoder(
    encoding_method="count",  # to do frequency ==> encoding_method='frequency'
    variables=["Neighborhood", "Exterior1st", "Exterior2nd"],
)

count_enc.fit(X_train)
```

```python
count_enc.encoder_dict_
```

```python
X_train = count_enc.transform(X_train)
X_test = count_enc.transform(X_test)

X_train.head()
```
